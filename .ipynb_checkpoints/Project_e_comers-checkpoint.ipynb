```python
# объединяем данные о заказах и о покупателях
customers_orders = customers_df.merge(orders_df, how='inner', on='customer_id')
```

```python
customers_orders.head()
```

```python
# добавляем новые столбцы месяц и год заказа (метод resample не использован, т.к. ядро зависало)
customers_orders['month'] = customers_orders.order_approved_at.dt.month
customers_orders['year'] = customers_orders.order_approved_at.dt.year
```

```python
# считаем количество заказов сделаных каждым уникальным покупателем в месяц по годам
customer_orders_per_month = (customers_orders.groupby(['year', 'month', 'customer_unique_id'], as_index=False)
                             .agg({'customer_id': 'count'})
                             .rename(columns={'customer_id': 'purchases'}))
```

```python
# добавляем стобец с точным количеством недель в месяце
def weeks_num(month):
    num_days=(31,28,31,30,31,30,31,31,30,31,30,31)
    weeks = num_days[int(month-1)] / 7
    return weeks

customer_orders_per_month['weeks_in_month'] = customer_orders_per_month.month.apply(weeks_num)
```

```python
# считаем среднее количество покупок в неделю
customer_orders_per_month['avg_purchases_per_week'] = customer_orders_per_month.purchases / customer_orders_per_month.weeks_in_month
```

```python
# Напиши функцию на python, позволяющую строить когортный анализ.
# В период с января по декабрь выяви когорту с самым высоким на 3-й месяц
```

```python
def user_retention(data_orders, data_customers, start, stop):
    # Функция расчета когортного анализа для user_retention
    # На вход получает:
    # data_orders - данные о заказах
    # data_customers - данные о покупателях
    # start - дата начала анализа в виде Год-Месяц-День в формате string
    # stop - дата окончания анализа в виде Год-Месяц-День в формате string
    
    # Переводим даты начала и конца анализа в флормат Datetime
    start_day = datetime.strptime(start, '%Y-%m-%d')
    stop_day = datetime.strptime(stop, '%Y-%m-%d')
    
    # Объединяем данные из исходных датафрэймов
    orders_with_unique_customers = data_orders.merge(data_customers, on='customer_id')
    
    # обрезаем данные по датам начала и конца
    orders_with_unique_customers = orders_with_unique_customers.query(
                                  'order_purchase_timestamp >= @start_day and order_purchase_timestamp <= @stop_day')
    
    # добавляем столбец с когортой
    cohorts = (
        orders_with_unique_customers.groupby('customer_unique_id', as_index=False)
        .agg({'order_purchase_timestamp': 'min'})
        .rename(columns={'order_purchase_timestamp': 'cohort'})
        )
    cohorts['cohort'] = pd.to_datetime(cohorts.cohort.dt.strftime('%Y-%m'))
    orders_with_unique_customers = orders_with_unique_customers.merge(cohorts, on='customer_unique_id', how='left')
    
    # добавляем столбец с номером месяца для retention
    orders_with_unique_customers['month_retention'] = (orders_with_unique_customers.order_purchase_timestamp.dt.month
                                                   - orders_with_unique_customers.cohort.dt.month)
    
    # создаем сводную таблицу с абсолютными значениями по retention
    cohorts_num = (
        orders_with_unique_customers.groupby(['cohort', 'month_retention'], as_index=False)
        .agg({'customer_unique_id': 'nunique'})
        .pivot(index='cohort', columns='month_retention', values='customer_unique_id')
        )
    
    # расчитываем относительные показатели retention
    retantion_table = cohorts_num.div((cohorts_num.iloc[:,0]), axis=0)
    
    return retantion_table
```

```python
user_retention_2017 = user_retention(orders_df, customers_df, start='2017-01-01', stop='2017-12-31')
```

```python
ur_style = (user_retention_2017
            .style
            .set_caption('User retention by cohort')
            .background_gradient(cmap='viridis')
            .highlight_null('white')
            .format("{:.2%}", na_rep=""))
ur_style
```

```python
# создадим датафрэйм с данными о количестве товаров в заказе и их стоимости
order_items_price = order_items_df.drop(columns=['product_id', 'seller_id', 'shipping_limit_date', 'freight_value'])
```

```python
orders_df.head(2)
```

```python
# создадим датафрэйм с данными о номере заказа, id покупателя, датой создания заказа и датой оплаты заказа
orders = orders_df.drop(columns=['order_status', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date'])
```

```python
customers_df.head(2)
```

```python
# создадим датафрэйм с данными об уникальных id клиентов
customers = customers_df.drop(columns=['customer_zip_code_prefix', 'customer_city', 'customer_state'])
```

```python
# объединим полученые датафрэймы в один
rfm_analis = customers.merge(orders, on='customer_id').merge(order_items_price, on='order_id')
# удалим лишние колонки, которые были необходимы для слияния
rfm_analis = rfm_analis.drop(columns=['customer_id','order_id'])
# посчитаем выручку для оплаченных заказов
rfm_analis['order_value'] = (rfm_analis[rfm_analis.order_approved_at.isna() != True].order_item_id 
                             * rfm_analis[rfm_analis.order_approved_at.isna() != True].price)

rfm_analis = rfm_analis.drop(columns=['order_item_id', 'price', 'order_approved_at'])
rfm_analis = rfm_analis.rename(columns={'order_purchase_timestamp': 'order_date'})

rfm_analis.head(3)
```

```python
# найдем дату самого последнего заказа
last_date = rfm_analis.order_date.max()
last_date
```

```python
# создадим сводную таблицу с дaнными об Recency, Frequency, Monetary для каждого клиента
rfmTable = (
            rfm_analis.groupby('customer_unique_id')
            .agg({'order_date': lambda x: (last_date - x.max()).days, # Recency #Количество дней с последнего заказа
                  'customer_unique_id': lambda x: len(x),      # Frequency #Количество заказов
                  'order_value': lambda x: x.sum()}) # Monetary Value #Общая сумма по всем заказам
            )

rfmTable['order_date'] = rfmTable['order_date'].astype(int)
rfmTable.rename(columns={'order_date': 'recency', 
                         'customer_unique_id': 'frequency', 
                         'order_value': 'monetary_value'}, inplace=True)
```

```python
rfmTable.head()
```

```python
# зададим функции для сегментации

quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])

def RClass(value,parameter_name,quantiles_table):
    if value <= quantiles_table[parameter_name][0.25]:
        return 1
    elif value <= quantiles_table[parameter_name][0.50]:
        return 2
    elif value <= quantiles_table[parameter_name][0.75]: 
        return 3
    else:
        return 4


def FMClass(value, parameter_name,quantiles_table):
    if value <= quantiles_table[parameter_name][0.25]:
        return 4
    elif value <= quantiles_table[parameter_name][0.50]:
        return 3
    elif value <= quantiles_table[parameter_name][0.75]: 
        return 2
    else:
        return 1
```

```python
# Проведем сегментацию

rfmSegmentation = rfmTable

rfmSegmentation['R'] = rfmSegmentation['recency'].apply(RClass, args=('recency',quantiles))

rfmSegmentation['F'] = rfmSegmentation['frequency'].apply(FMClass, args=('frequency',quantiles))

rfmSegmentation['M'] = rfmSegmentation['monetary_value'].apply(FMClass, args=('monetary_value',quantiles))

rfmSegmentation['RFMClass'] = rfmSegmentation.R.map(str) \
                              + rfmSegmentation.F.map(str) \
                              + rfmSegmentation.M.map(str)
```

```python
rfmSegmentation.head()
```

```python
# Создадим датафрэйм с данными о средних значениях по каждому показателю для каждого кластера
avg_clasters_values = pd.DataFrame()
for i in [1,2,3]:
    for j in [1,2,3]:
        for k in [1,2,3]:
            claster = str(i) + str(j) + str(k)
            mean_recency = rfmSegmentation.query('R == @i' and 'F == @j' and 'M == @k').recency.mean()
            mean_frequency = round(rfmSegmentation.query('R == @i' and 'F == @j' and 'M == @k').frequency.mean())
            mean_money = round(rfmSegmentation.query('R == @i' and 'F == @j' and 'M == @k').monetary_value.mean(), 2)
            rows_values = [claster, mean_recency, mean_frequency, mean_money]
            avg_clasters_values = avg_clasters_values.append([rows_values], ignore_index=True)

avg_clasters_values.columns = ['claster', 'mean_recency', 'mean_frequency', 'mean_money']
            
```
